```python
gdf_web_mercator = gdf.to_crs(epsg=3857)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))

gdf_web_mercator[gdf_web_mercator['Median Income'] < 57473].plot(ax=ax, alpha=0.8)

# turn the axis off
ax.axis('off')

# set a title
ax.set_title('Census Tracts with less than $57,473 Median Income',fontsize=16)

# add a basemap
ctx.add_basemap(ax)
```
