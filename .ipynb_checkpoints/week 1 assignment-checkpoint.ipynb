## Creating new variable for types of cities.
###### The goal is to create a new variable with specific data we can use to plote into a chart. This will provide a quick visualizaton of the total number of cities that are in each category. In order to successfully pull the correct data we need to use "Type" column as indicated below.

```python
citytype_count = neighborhood['type'].value_counts()
citytype_count
```

## This allows us to reset the name of the columns and make them more tailored to the names that we want on our charts and plots.

```python
citytype_count = citytype_count.reset_index()
citytype_count
```

## Type of Data for our new variable

```python
type(citytype_count)
```

## Renaming Columns for our 2 lists

```python
citytype_count.columns.to_list()
```

## New names being used for 2 columns in new variable dataset

```python
citytype_count.columns = ['type', 'total']
```

```python
citytype_count
```

## Quick plot of our data point.
###### This plot does not make much sense to use. It should not have a linear relationship, but instead one where we can see the differences between city types.

```python
citytype_count.plot()
```

## Bar Chart Plot

```python
citytype_count.plot.bar(x = 'type', y = 'total', title = 'Types of City in Los Angeles')
```

```python
## Desired Columns - An alternative code that can be used.
```

```python
desired_columns = ['name','type','geometry']
neighborhood[desired_columns]
```

## Creating a new variable of the desired data
###### This will make sure we only use data we want and not the entire dataset. This can make things messy.

```python
neighborhood_trimmed = neighborhood[desired_columns].copy()
neighborhood_trimmed
```

## Double check data is correct in new datset

```python
neighborhood_trimmed.head()
```

## Specific city in dataset and type of city in county

```python
neighborhood_trimmed[neighborhood_trimmed.name == 'Alhambra']
```

## An Alternative Way

```python
# another way .loc
neighborhood_trimmed.loc[neighborhood_trimmed['type'] == 'standalone-city']
```

## More ways this time using the type to pull up citites in its category

```python
# another way .query
neighborhood_trimmed.query("type == 'unincorporated-area'")
```

## Plot of LA County
